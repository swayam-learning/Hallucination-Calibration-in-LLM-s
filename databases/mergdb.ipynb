```python
import sqlite3
import pandas as pd

# ===============================
# DATABASE PATHS
# ===============================

db_paths = [
    r"d:\Hallucination Calibration In LLM's\databases\gen_1B.db",
    r"d:\Hallucination Calibration In LLM's\databases\gen_3B.db",
    r"d:\Hallucination Calibration In LLM's\databases\gen_7B.db"
]


# ===============================
# LOAD AND MERGE DATABASES
# ===============================

dfs = []

for path in db_paths:
    print(f"Loading: {path}")

    conn = sqlite3.connect(path)
    df = pd.read_sql("SELECT * FROM generations", conn)
    conn.close()

    dfs.append(df)

# Merge all dataframes
merged_df = pd.concat(dfs, ignore_index=True)

print("\nMerged dataset shape:", merged_df.shape)

# ===============================
# SAVE MERGED DATASET
# ===============================

merged_df.to_csv("d:\Hallucination Calibration In LLM's\databases\merged_generations.csv", index=False)

print("Merged dataset saved to databases/merged_generations.csv")

# ===============================
# QUICK CHECK
# ===============================

print("\nRows per model:")
print(merged_df["model_name"].value_counts())

print("\nRows per band:")
print(merged_df["band"].value_counts())
```

```python
import pandas as pd

# 1 Load your merged dataset
df = pd.read_csv(r"D:\Hallucination Calibration In LLM's\databases\merged_generations.csv")

# 2 Load TruthfulQA
truth = pd.read_csv(r"D:\Hallucination Calibration In LLM's\data\data\original\TruthfulQA.csv")

# 3 Merge
df = df.merge(
    truth[["Question","Correct Answers","Incorrect Answers"]],
    left_on="question",
    right_on="Question",
    how="left"
)

# 4 Remove duplicate Question column (optional but clean)
df = df.drop(columns=["Question"])

# 5 Save new dataset
df.to_csv(
    r"D:\Hallucination Calibration In LLM's\databases\merged_with_truthfulqa.csv",
    index=False
)

print("New dataset saved successfully")
print(df.shape)
```
